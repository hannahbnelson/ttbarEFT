```python
def plot_single_hist(hist):
    fig, ax = plt.subplots(1,1)
    hist.plot1d(histtype="step", ax=ax, stack=False, density=True)
    ax.legend(loc='upper right', fontsize=14)
    plt.figtext(0.13, 0.89, "Reweighted to SM", fontsize=25)
    return fig
```

```python
def plot_two_hist(hist1, hist2):
    fig, ax = plt.subplots(1,1)
    hist1.plot1d(histtype="step", ax=ax, stack=False, density=True)
    hist2.plot1d(histtype="step", ax=ax, stack=False, density=True)
    ax.legend(loc='upper right', fontsize=14)
    plt.figtext(0.13, 0.89, "Reweighted to SM", fontsize=25)
    return fig
```
