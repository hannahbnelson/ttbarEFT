```python
import numpy as np
import awkward as ak

import hist
from topcoffea.modules.histEFT import HistEFT
import topcoffea.modules.utils as utils

import mplhep as hep
import matplotlib.pyplot as plt
```

```python
hep.style.use("CMS")
params = {'axes.labelsize': 20,
          'axes.titlesize': 20,
          'legend.fontsize':20}
plt.rcParams.update(params)
```

```python
fin = "../central_ttbarEFT_test.pkl.gz"
hists = utils.get_hist_from_pkl(fin, allow_empty=False)
print(hists)
```

```python
h = hists['njets']
```

```python
h.axes[0]
```

```python
h.wc_names
```

```python
h_rwgt = h.as_hist({})
```

```python
fig, ax = plt.subplots(1,1)
h_rwgt.plot1d(histtype="step", ax=ax, stack=False)
ax.legend(loc='upper right', fontsize=14)
```

```python
fig, ax = plt.subplots(1,1)
hep.histplot(h_rwgt.project("njets"), histtype='fill', label="TT01j2l")
ax.legend(loc='upper right', fontsize=14)
```
