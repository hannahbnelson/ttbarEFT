```python
hep.style.use("CMS")
params = {'axes.labelsize': 20,
          'axes.titlesize': 20,
          'legend.fontsize':20}
plt.rcParams.update(params)
```

```python
f_EFT = "../hists/central_ttbarEFT.pkl.gz"
hists_EFT = utils.get_hist_from_pkl(f_EFT, allow_empty=False)

f_SM = "../hists/central_ttbarPowheg.pkl.gz"
hists_SM = utils.get_hist_from_pkl(f_SM, allow_empty=False)
```

```python
def plot_single_hist(hist):
    fig, ax = plt.subplots(1,1)
    hist.plot1d(histtype="step", ax=ax, stack=False, density=True)
    ax.legend(loc='upper right', fontsize=14)
    plt.figtext(0.13, 0.89, "Reweighted to SM", fontsize=25)
    return fig
```

```python
def plot_two_hist(hist1, hist2):
    fig, ax = plt.subplots(1,1)
    hist1.plot1d(histtype="step", ax=ax, stack=False, density=True)
    hist2.plot1d(histtype="step", ax=ax, stack=False, density=True)
    ax.legend(loc='upper right', fontsize=14)
    plt.figtext(0.13, 0.89, "Reweighted to SM", fontsize=25)
    return fig
```

```python
# for name in hists_EFT.keys(): 
#     h_rwgt = hists_EFT[name].as_hist({})
#     fig = plot_hist_sm(h_rwgt)
#     fig.show()

for name in hists_EFT.keys():
    h_EFT = hists_EFT[name].as_hist({})
    h_SM = hists_SM[name].as_hist({})
    fig = plot_two_hist(h_EFT, h_SM)
    plt.savefig("centralttbarpowheg_"+name+".png")
```
